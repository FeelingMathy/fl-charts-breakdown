```python
#import libraries for analysis and data viz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

%matplotlib inline

#read in file, change directory to flavor
xlsx = pd.ExcelFile('X:\\FlHealthCharts\\frequency and count of r99 deaths by county and year.xlsx')
#all goodies are on one sheet
df = pd.read_excel(xlsx, 'analysis')
```

```python
#pulling X axis index of county names
counties = df.loc[:,'County Name']

#pulling Y_1 axis column of average monthly deaths 1999-2019 from FDOH vital statistics into a new dataframe
avg_1999_2019 = df.loc[0:67, '1999-2019 average monthly deaths (all causes)']
#pulling Y_2 axis column of average monthly deaths 2020 (provisional, through June) from FDOH vital statistics into a new dataframe
avg_2020 = df.loc[0:67, '2020 average monthly deaths (provisional, all causes)']

```

```python
#adds grid behind plot
plt.style.use('ggplot')

#sets plot size to something readable
plt.figure(figsize=(20,10))

#sets axis labels and title
x_pos = [i for i, _ in enumerate(counties)]

plt.xlabel("County")
plt.ylabel("Average monthly deaths, all causes")
plt.title("Average monthly deaths by county for all causes, 1999-2019 and 2020 (provisional)")

#sets tick marks
plt.xticks(x_pos, counties,rotation='vertical')

#plots as bar chart with error bars
plt.bar(x_pos, avg_2020, color='yellow',width = 0.5, label = "Monthly deaths, all causes, 2020 (provisional)",edgecolor='black')
plt.bar(x_pos, avg_1999_2019, color='green',width = 0.5, label = "Monthly deaths, all causes, 1999-2019",edgecolor='black')

#creates legend for data
plt.legend(loc='best')

#saves graph
plt.savefig('X:\\FlHealthCharts\\average monthly deaths all causes 1999_2020.png')
#shows plot
plt.show()

```
